## Spotify Veri Analizi Projesi

Bu proje, Spotify'dan alınan bir veri seti kullanılarak müzik trendlerini analiz etmeyi amaçlamaktadır. Proje, çeşitli şarkı özelliklerinin analiz edilmesini ve görselleştirilmesini içerir.

## Senorya
Bir dijital müzik platformunda çalıştığınızı hayal edin. Şirket, kullanıcıların dinleme alışkanlıklarını anlamak ve gelecekteki trendleri tahmin ederek daha iyi öneriler sunmak için bir analiz yapılmasını istiyor. Spotify veri setini kullanarak, platformda öne çıkan özellikleri incelemek ve stratejik öneriler geliştirmek sizin göreviniz.
Problem Tanımı

## Problem Tanımlaması
Müzik endüstrisi, dinleyici tercihlerini anlamak ve değişen trendleri yakalamak için büyük veri analizine giderek daha fazla ihtiyaç duymaktadır. Spotify gibi dijital müzik platformları, kullanıcıların dinleme alışkanlıklarından elde edilen büyük miktarda veriyle çalışır. Ancak bu verilerden anlamlı içgörüler çıkarmak, etkili stratejiler oluşturmak için kritik bir öneme sahiptir.

Bu bağlamda şu sorulara yanıt aranması gerekmektedir:

1. Popüler Şarkılar: Popüler şarkıların özellikleri nelerdir ve bu özellikler şarkıların başarısını nasıl etkiler?


2. Tür Trendleri: Belirli müzik türlerinin zaman içindeki popülerliği nasıl değişiyor ve bu değişimi etkileyen faktörler nelerdir?


3. Korelasyonlar: Şarkı özellikleri (dans edilebilirlik, enerji, tempo vb.) arasında anlamlı bir ilişki var mı ve bu ilişkiler dinleyici tercihlerini nasıl şekillendiriyor?


4. Öngörüler: Gelecekte hangi şarkı özellikleri ve türler daha popüler olabilir? Bu öngörüler nasıl stratejik avantaj sağlar?



## Hedef

Spotify veri seti kullanılarak, şarkıların özellikleriyle popülerlikleri arasındaki ilişkiyi belirlemek, türlerin zaman içindeki değişimini analiz etmek ve geleceğe yönelik öneriler geliştirmek. Bu analiz, müzik platformlarının içerik oluşturma ve kullanıcıya özel öneriler geliştirme stratejilerine rehberlik edecek içgörüler sunacaktır.

Bu problem, veri analitiği ve görselleştirme araçları kullanılarak çözülebilir ve müzik endüstrisi için değerli çıkarımlar elde edilmesini sağlar.

## Veri Seti Özellikleri

Spotify veri seti, müzik şarkılarının popülerliğini ve özelliklerini anlamak için çeşitli sütunlar içerir. Veri setindeki temel özellikler şunlardır:

Track Name (Şarkı Adı): Şarkının adı.

Artist Name (Sanatçı): Şarkıyı icra eden sanatçının adı.

Genre (Tür): Şarkının müzik türü (örneğin pop, rock, hip-hop).

Release Date (Yayınlanma Tarihi): Şarkının piyasaya sürüldüğü tarih.

Popularity (Popülerlik): 0-100 arasında değişen bir skor; şarkının Spotify üzerindeki popülerliğini gösterir.

Danceability (Dans Edilebilirlik): 0.0 ile 1.0 arasında değişen bir skor; şarkının dans edilebilirliğini ifade eder.

Energy (Enerji): 0.0 ile 1.0 arasında bir değer; şarkının yoğunluğunu ve aktivite seviyesini tanımlar.

Acousticness (Akustiklik): Şarkının ne kadar akustik olduğuna dair bir ölçüm.

Tempo (Tempo): Şarkının temposunu (BPM olarak) gösterir.

Duration (Süre): Şarkının süresi (milisaniye olarak).




## Kullanılan Kütüphaneler ve Amaçları

Projede, veri analizi ve görselleştirme işlemleri için aşağıdaki Python kütüphaneleri kullanılmıştır:

1. Pandas:

Veri yükleme, temizleme ve işleme işlemleri için kullanıldı.

Örnek: Eksik değerlerin kontrol edilmesi, belirli şarkıların filtrelenmesi ve özet istatistiklerin çıkarılması.




2. NumPy:

Matematiksel işlemler ve veri manipülasyonu için kullanıldı.

Örnek: Popülerlik skorlarının normalleştirilmesi veya ortalama değerlerin hesaplanması.



3. Matplotlib:

Temel görselleştirme araçları için kullanıldı.

Örnek: Popülerlik dağılımını histogram olarak gösterme.




4. Seaborn:

İleri düzey ve estetik görselleştirmeler için kullanıldı.

Örnek: Dans edilebilirlik ve enerji arasındaki ilişkiyi incelemek için scatter plot (dağılım grafiği) oluşturma.




5. Scikit-learn:

Basit makine öğrenmesi modelleri ve veri kümelerini ayırmak için kullanıldı.

Örnek: Popülerliği tahmin etmek için regresyon modeli oluşturma.





## Yapılan Analizler

1. Tanımlayıcı İstatistikler:
Veri setinin genel yapısını anlamak için sütunların ortalama, medyan, minimum ve maksimum değerleri hesaplandı.


2. Korelasyon Analizi:
Şarkı özellikleri arasındaki ilişkiler (örneğin enerji ve dans edilebilirlik) korelasyon katsayıları ile incelendi.


3. Trend Analizi:
Zaman içinde farklı türlerin popülerlik değişimi incelendi.


4. Görselleştirme:

Histogramlar: Popülerlik ve tempo gibi özelliklerin dağılımı.

Çizgi Grafikleri: Türlerin zaman içindeki değişimi.

Isı Haritaları: Korelasyonların görselleştirilmesi.



5. Tahmin Modelleri:
Popülerlik skorunu tahmin etmek için basit lineer regresyon denemeleri yapıldı.



Bu kütüphaneler ve analizler, müzik verilerinden anlamlı içgörüler çıkarılmasını ve görsel olarak raporlanmasını sağladı.
